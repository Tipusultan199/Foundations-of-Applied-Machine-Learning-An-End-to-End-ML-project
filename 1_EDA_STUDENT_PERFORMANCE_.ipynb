### Outlier Detection using IQR

$$
Q1 = 25^{th}\ \text{percentile}, \quad Q3 = 75^{th}\ \text{percentile}
$$

$$
IQR = Q3 - Q1
$$

$$
\text{Lower Bound} = Q1 - 1.5 \times IQR
$$

$$
\text{Upper Bound} = Q3 + 1.5 \times IQR
$$

$$
\text{Outlier if } \; x < \text{Lower Bound} \; \text{ or } \; x > \text{Upper Bound}
$$
The 1.5 in the IQR formula is a rule of thumb multiplier.

```python
# --- Function to remove outliers using IQR ---
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

# --- Apply for multiple columns ---
clean_df = df.copy()
for col in ['math score', 'reading score', 'writing score', 'average']:
    clean_df = remove_outliers(clean_df, col)

# --- Plot after removing outliers ---
plt.subplots(1, 4, figsize=(16, 5))

plt.subplot(141)
sns.boxplot(x=clean_df['math score'], color='skyblue')
plt.title("Math Score")

plt.subplot(142)
sns.boxplot(x=clean_df['reading score'], color='hotpink')
plt.title("Reading Score")

plt.subplot(143)
sns.boxplot(x=clean_df['writing score'], color='yellow')
plt.title("Writing Score")

plt.subplot(144)
sns.boxplot(x=clean_df['average'], color='lightgreen')
plt.title("Average Score")

plt.tight_layout()
plt.show()
```
